# Business Profit Analyzer

**INTRODUCTION**
- “Business Profit Leakage Analyzer is a data-driven system designed to identify, monitor, and reduce revenue losses.” 
- Means:-ek smart system hai jo business ka data check karega aur batayega:“Company ka paisa kaha loss ho raha hai?”

**Profit Leakage**
- 👉 Jab business:
- unnecessary discount deta hai
- products return hote hain
- stock waste hota hai
- pricing galat hoti hai
- ➡ company ko hidden loss hota hai.👉 Is hidden loss ko:🔥 Profit Leakage kehte hain.
  
**There is 5 leakage sources given in documentation:-**
- Discount misuse:-zyada discount
- Returns & refunds:-products wapas aana
- Inventory wastage:-stock damage/loss
- Pricing inconsistency:-wrong price
- Operational inefficiency:-process problems

**🎯 Main Goal:-**
- ✔ hidden loss detect karega
- ✔ management ko alert karega



# Summary of the dataset

**Superstore Dataset**

The Sample Superstore Dataset is a retail business dataset that contains information about customer orders, product sales, discounts, profits, shipping details, and regional performance. This dataset represents the day-to-day transactions of a superstore business where different products are sold to customers across multiple cities and regions.

Each row in the dataset represents a single customer order or transaction. The dataset helps in understanding how the business is performing financially and operationally.

The dataset contains around 9994 rows and 13 columns and is clean and ready for analysis.

One important observation from the dataset is that high discounts often lead to low profit or loss. This makes the dataset suitable for building a Business Profit Leakage Analyzer project.

**Column Description**

- Ship Mode-Type of delivery used for shipping
- Segment	-Type of customer segment (Consumer, Corporate, Home Office)
- Country-Country where sales happened
- City-City of the customer/order
- State-State where order was placed
- Postal Code-Postal code of the area
- Region-Business region (East, West, Central, South)
- Category-Main product category
- Sub-Category-Product sub-category
- Sales-Total sales amount
- Quantity-Number of items sold
- Discount-Discount given on the product
- Profit-Profit earned or loss suffered

**Additional Insights**

- The dataset shows that high discounts often reduce profit and sometimes create losses.
- Some product categories and regions perform better than others in terms of profit.
- Even when sales are high, profit can still be low because of excessive discounts.

# FEATURE ENGINEERING

```python
# Profit Status Column:-Har transaction ko classify karna:
store['Profit_Status'] = store['Profit'].apply(lambda x: 'Loss' if x < 0 else 'Profit')
```

```python
store.head()
```

```python
# Loss Flag Column:-Har transaction ko:numeric flag dena.Loss=1 and profit=0
store['Loss_flag']=store['Profit'].apply(lambda x: 1 if x<0 else 0) 
```

```python
store.head()
```

```python
# High Discount Flag:-Har transaction ko identify karna:High discount hai ya nahi.(30% se jyada then 1 and 30% se kam then 0)
store['High_Discount_Flag'] = store['Discount'].apply(lambda x: 1 if x > 0.3 else 0)
```

```python
store.head()
```

```python
# Discount Percentage Column:-
store["Discount_percentage"]=store['Discount']*100
```

```python
store.head()
```

```python
# Profit Margin Column:-Sales ke comparison me kitna profit hua hai
store['Profit_margin_percentage']=round((store['Profit']/store['Sales'])*100,2)
```

```python
store.head()
```

```python
store.head()
```

```python
# Risk Category:-Har transaction ko:risk level dena.
store['Risk_Category'] = store.apply(
    lambda row: 'High Risk' if row['Profit'] < 0 and row['Discount'] > 0.3
    else ('Medium Risk' if row['Profit'] < 0
    else 'Low Risk'),
    axis=1
)
```

# ML


**Isolation Forest**

```python
# In this project we use Isolation Forest because anomaly detection ke liye best hai.
#anomaly detection means:Normal data se alag unusual / suspicious data
# Goal:-Business data me:suspicious / risky transactions automatically detect karna Using:Isolation Forest.
```

```python
from sklearn.ensemble import IsolationForest
```

```python
IsolationForest
```

```python
# Feature Selection:-
# Goal:-Model ko:kaunse columns se learning karani hai wo select karna
# Sirf important numeric columns ML ke liye select karo
features=store[['Sales','Profit', 'Discount','Quantity']]
```

```python
features.head()
```

```python
# Create & Train Model:-
# Hum:Isolation Forest model banayenge aur Ye model:suspicious transactions detect karega
model = IsolationForest(contamination=0.05, random_state=42)
model.fit(features)
```

```python
# Predict Anomalies:-
# Goal:-Model se puchna:Kaunsi transactions normal hain aur kaunsi suspicious.
store['Anomaly'] = model.predict(features)
```

```python
# Verify anomalies:-
store[['Sales',
       'Profit',
       'Discount',
       'Anomaly']].head(10)
```

```python
#Count Total Anomalies:-
# Goal:-Find karna:-Model ne kitni suspicious transactions detect ki hain.

store['Anomaly'].value_counts()

#Model detected around 499 potentially risky or abnormal business transactions
```

```python
# Extract Suspicious Transactions:-Sirf:suspicious rows dekhna
anomalies=store[store['Anomaly']==-1]
```

```python
anomalies.head()
```

```python
# Analyze Anomalies:-ML ne jo suspicious transactions detect ki hain,unme kya patterns hain
# FIRST ANALYSIS:-Average Discount in Anomalies

round(anomalies['Discount'].mean()*100,2)

#Model detected that suspicious transactions generally contain very high discounts.
#Suspicious transactions contain an average discount of approximately 43%.
```

```python
# Average Profit in Anomalies:-ML-detected suspicious transactions profit de rahi hain ya loss

round(anomalies['Profit'].mean(), 2)

#ML-detected anomalies: average me:profit de rahi hain
# that's mean Har anomaly loss nahi hoti.
#Isolation Forest:sirf:losses nahi dekhta Wo:unusual patterns detect karta hai
#The anomaly model detected both high-risk loss transactions and unusually large profitable transactions.
```

```python
# Visualize Anomalies:-Normal vs suspicious transactions
plt.figure(figsize=(10,6))

sns.scatterplot(x=store['Sales'],
                y=store['Profit'],
                hue=store['Anomaly'])

plt.title('Anomaly Detection')
plt.show()

#The model identified isolated transactions with extreme sales and profit/loss values as anomalies.
#Most normal transactions are concentrated around moderate sales and profit values,while suspicious transactions appear isolated with extreme business behavior.



```

```python
#Final ML Insights:-
print("Total Suspicious Transactions:",
      (store['Anomaly'] == -1).sum())

print("Average Discount in Anomalies:",
      round(anomalies['Discount'].mean()*100,2), "%")

print("Average Profit in Anomalies:",
      round(anomalies['Profit'].mean(),2))
```

**Random Forest**

```python
#Random Forest(Prediction Model):-model predict karega:Transaction profit me jayega ya loss me
from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
```

```python
#Features & Target Selection:-Input kya hoga aur output kya predict karna hai
#Random Forest Me 2 Important Things Hoti Hain:-
# 1.Features (X)-input columns(Sales,Quantity,Discount,Profit_margin_percentage)
# 2.Target (y)-prediction column(Profit_Status)
x = store[['Sales','Quantity', 'Discount']]
y = store['Profit_Status']
```

```python
x.head()
```

```python
y.head()
```

```python
#Train-Test Split:-
#Dataset ko:2 parts me divide karna
# 1.training Data-model ko sikhana
# 2.Testing Data-model ko check karna
x_train, x_test, y_train, y_test = train_test_split(x,y,
    test_size=0.2,
    random_state=42
)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#Train Random Forest Model:-Random Forest ko:training data se learning karwana

rf_model = RandomForestClassifier(random_state=42)

rf_model.fit(x_train, y_train)

```

```python
# Make Predictions:-Testing data pe:model se:predictions karwana
y_pred = rf_model.predict(x_test)
```

```python
y_pred[:10]
```

```python
#Accuracy Check:-Model kitna sahi predict kar raha hai
accuracy = accuracy_score(y_test, y_pred)

print("Model Accuracy:", accuracy)

#The Random Forest model achieved approximately 94% prediction accuracy in identifying profit and loss transactions.
```

```python
#Random Forest Visualization:-Kaunsa feature prediction me sabse important hai.
importance = rf_model.feature_importances_

feature_names = x.columns
```

```python
plt.figure(figsize=(8,5))

sns.barplot(x=importance,y=feature_names)
            
plt.title('Feature Importance - Random Forest')

plt.xlabel('Importance Score')
plt.ylabel('Features')

plt.show()

#High discounts are the strongest indicator of business losses and risky transactions.
#Sales bhi impact karta hai but discount jitna nahi
#Quantity ka impact profit/loss prediction me kam hai
```

**K-Means Clustering**

```python
#K-Means Clustering:- Grouping/segmentation algorithm import karna
#K-Means Clustering:-Ye:similar transactions ko:groups/clusters me divide karega
#K-Means groups similar transactions together based on business behavior patterns.
from sklearn.cluster import KMeans
```

```python
#Select Features for Clustering:-Kin columns ke basis pe groups banane hain
cluster_features = store[['Sales','Profit','Discount']]
```

```python
cluster_features.head()
```

```python
#Create K-Means Model:Kitne groups/clusters banane hain
kmeans = KMeans(n_clusters=3, random_state=42)
#K-Means will divide transactions into 3 behavior-based groups.
```

```python
#Train K-Means Model:-Model ko:transaction patterns sikhana
kmeans.fit(cluster_features)
```

```python
#Assign Cluster Labels:-Assign Cluster Labels cluster number assign karna
store['Cluster'] = kmeans.labels_
```

```python
store[['Sales',
       'Profit',
       'Discount',
       'Cluster']].head(10)
```

```python
#Analyze Cluster Distribution:-Har cluster me kitni transactions hain
store['Cluster'].value_counts()
```

```python
#Understand Cluster Behavior:-Har cluster ka behavior kaisa hai
store.groupby('Cluster')[['Sales','Profit', 'Discount']].mean()
```

```python
#K-Means Visualization:-Different transaction groups
plt.figure(figsize=(10,6))

sns.scatterplot(x=store['Sales'],
                y=store['Profit'],
                hue=store['Cluster'],
                palette='Set1')

plt.title('K-Means Clustering')

plt.show()
```
